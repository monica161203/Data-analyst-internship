```python
import easyocr
import pandas as pd
import re
from pdf2image import convert_from_path
from PIL import Image

# Replace with your uploaded file name
pdf_path = "6.pdf"
pages = convert_from_path(pdf_path)

reader = easyocr.Reader(['en'])
all_text_lines = []

# OCR all pages
for i, page in enumerate(pages):
    image_path = f"/content/page_{i+1}.jpg"
    page.save(image_path, "JPEG")
    text = reader.readtext(image_path, detail=0)
    all_text_lines.extend(text)

# Join text for regex search
full_text = "\n".join(all_text_lines)

# Helper function to extract via regex
def extract_with_regex(pattern, text, group=1):
    match = re.search(pattern, text, re.IGNORECASE)
    return match.group(group).strip() if match else ""

# Field mappings and patterns
invoice_data = {
    "Bill Number": extract_with_regex(r'Bill\s*(No|Number)[^\w]?\s*[:\-]?\s*([A-Z0-9\-\/]+)', full_text, 2),
    "Patient Name": extract_with_regex(r'Patient\s*Name[^\w]?\s*[:\-]?\s*([A-Z ]+)', full_text, 1),
    "Bill Date": extract_with_regex(r'(\d{2}/\d{2}/\d{4}.*?(?:AM|PM))', full_text, 1),
    "Bill Type": extract_with_regex(r'Bill\s*Type[^\w]?\s*[:\-]?\s*(\w+)', full_text, 1),
    "Consultant": extract_with_regex(r'Consultant[^\w]?\s*[:\-]?\s*([A-Z ]+)', full_text, 1),
    "OP No": extract_with_regex(r'OP\s*No[^\w]?\s*[:\-]?\s*(\S+)', full_text, 1)
}

# Save to DataFrame
invoice_df = pd.DataFrame([invoice_data])
invoice_df.to_csv("invoice_output.csv", index=False)

# Download the CSV
from google.colab import files
files.download("invoice_output.csv")
```

```python
import easyocr, torch, re
from pdf2image import convert_from_path
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
import numpy as np # Import numpy

reader = easyocr.Reader(['en'])
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-handwritten")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def run_trocr(image):
    pixel_values = processor(images=image, return_tensors="pt").pixel_values.to(device)
    generated_ids = model.generate(pixel_values)
    return processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

def extract_all_text(pdf_path):
    pages = convert_from_path(pdf_path)
    for i, page in enumerate(pages):
        print(f"\n--- Page {i+1} ---")
        # Convert PIL Image to NumPy array
        page_np = np.array(page)
        results = reader.readtext(page_np) # Pass the NumPy array to readtext
        all_fields = []
        for bbox, text, conf in results:
            x_min = min(p[0] for p in bbox)
            y_min = min(p[1] for p in bbox)
            x_max = max(p[0] for p in bbox)
            y_max = max(p[1] for p in bbox)
            pad = 5
            crop_box = (max(0, x_min - pad), max(0, y_min - pad), x_max + pad, y_max + pad)
            # Crop the original PIL Image, not the NumPy array
            field_img = page.crop(crop_box)
            trocr_text = run_trocr(field_img)
            all_fields.append(trocr_text)
        for line in all_fields:
            print(line)

extract_all_text("2____New_Invoices.pdf")
```

```python
import pdfplumber
import pandas as pd
import re

#  STEP 1: Define PDF path
pdf_path = "Jennifer.Wilson.Hospital.Charges.09.26.2024 Dhivya jamil-5-2.pdf"

#  STEP 2: Extract patient blocks using DOB pattern
def extract_blocks_from_text(text):
    lines = text.split("\n")
    blocks = []
    current_block = []
    for line in lines:
        if re.search(r"\b\d{1,2}/\d{1,2}/\d{4}\b", line):  # line with DOB
            if current_block:
                blocks.append(" ".join(current_block))
                current_block = []
        current_block.append(line.strip())
    if current_block:
        blocks.append(" ".join(current_block))
    return blocks

#  STEP 3: Extract fields using regex
def extract_patient_info(block_text):
    name = re.search(r"[A-Z][a-z]+, [A-Z][a-z]+(?: [A-Z])?", block_text)
    dob = re.search(r"\b\d{1,2}/\d{1,2}/\d{4}\b", block_text)
    mrn = re.search(r"\b\d{6,8}\b", block_text)
    unit = re.search(r"(BCH MEDICAL SURGICAL|CFV \d+ [A-Z]+\s*\d*[A-Z]*)", block_text)
    room = re.search(r"\d{3}[A-Z]?/\d{3}[A-Z]?-?\d{2}", block_text)
    problem = re.search(r"([A-Za-z0-9 ,/\-\(\)]+?)\s*\(Principal Hospital Problem\)", block_text)
    payor = re.search(r"(20\d{2}|1247|1085|1031|1001|308013|3002|H084)[- ]+([A-Z][A-Z\s]+)", block_text.upper())

    return {
        "Patient Name": name.group(0) if name else "",
        "DOB": dob.group(0) if dob else "",
        "MRN": mrn.group(0) if mrn else "",
        "Unit": unit.group(0).strip() if unit else "",
        "Room/Bed": room.group(0) if room else "",
        "Primary Problem": problem.group(1).strip() if problem else "",
        "Primary Payor": f"{payor.group(1)} - {payor.group(2).strip()}" if payor else ""
    }

#  STEP 4: Load PDF and apply logic
with pdfplumber.open(pdf_path) as pdf:
    all_text = "\n".join(page.extract_text() for page in pdf.pages if page.extract_text())

blocks = extract_blocks_from_text(all_text)
patient_records = [extract_patient_info(block) for block in blocks if extract_patient_info(block)["Patient Name"]]

#  STEP 5: Convert to DataFrame and Save
df = pd.DataFrame(patient_records)
df.to_csv("extracted_patients.csv", index=False)
df.to_excel("extracted_patients.xlsx", index=False)

print("Extraction complete. Files saved: extracted_patients.csv, extracted_patients.xlsx")
```

```python
import camelot
import pandas as pd
from PyPDF2 import PdfReader
from google.colab import files

#  Step 1: Upload your PDF in Google Colab
uploaded = files.upload()
pdf_file = list(uploaded.keys())[0]

#  Step 2: Find the number of pages
reader = PdfReader(pdf_file)
total_pages = len(reader.pages)
print(f" Total pages in PDF: {total_pages}")
#  Step 3: Function to clean multiline rows
def clean_multiline_rows(df):
    cleaned_rows = []
    buffer_row = None

    for idx, row in df.iterrows():
        if pd.isnull(row[0]) or row[0].strip() == "":
            if buffer_row is not None:
                buffer_row = [f"{old} {new}".strip() for old, new in zip(buffer_row, row)]
        else:
            if buffer_row:
                cleaned_rows.append(buffer_row)
            buffer_row = row.tolist()
    if buffer_row:
        cleaned_rows.append(buffer_row)

    return pd.DataFrame(cleaned_rows, columns=df.columns)

#  Step 4: Extract and clean tables from each page
all_tables = []

for page_num in range(1, total_pages + 1):
    print(f" Extracting Page {page_num}...")
    tables = camelot.read_pdf(
        pdf_file,
        pages=str(page_num),
        flavor='stream',
        edge_tol=200,
        strip_text='\n'
    )

    if tables.n > 0:
        for table in tables:
            cleaned_df = clean_multiline_rows(table.df)
            all_tables.append(cleaned_df)
        print(f" Found {tables.n} table(s) on Page {page_num}")
    else:
        print(f" No tables found on Page {page_num}")

#  Step 5: Merge and save all extracted tables
if all_tables:
    merged_df = pd.concat(all_tables, ignore_index=True)
    merged_df.to_csv("merged_output.csv", index=False)
    print(" All tables saved into 'merged_output.csv'")
    files.download("merged_output.csv")
else:
    print(" No tables extracted from the PDF.")
```
